# Strategic Scenario Modeling
## Business Question

## Where should leadership focus investments?

## Import Required Libraries

```python
# import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyarrow
```

## Load Data

```python
# load data
df = pd.read_parquet("fmcg_cleaned.parquet")
```

## Scenario 1: 5% Price Increase


```python
## price increase
df['price_5pct_up'] = df['price_unit'] * 1.05
df['revenue_price_up'] = df['price_5pct_up'] * df['units_sold']

```

## Scenario 2: Increased Promotion Intensity

```python
## increased promotion
promo_push = df.copy()
promo_push.loc[promo_push['promotion_flag']==1, 'units_sold'] *= 1.1
promo_push['revenue'] = promo_push['units_sold'] * promo_push['price_unit']



```

## Scenario Comparison

```python
# scenario comparison
summary = pd.DataFrame({
    'Scenario': ['Base','Price +5%','Promo Push'],
    'Revenue': [
        df['revenue'].sum(),
        df['revenue_price_up'].sum(),
        promo_push['revenue'].sum()
    ]
})

summary
```

## Insights
### Pricing delivers cleaner revenue gains than promotion-heavy strategies under supply constraints.
