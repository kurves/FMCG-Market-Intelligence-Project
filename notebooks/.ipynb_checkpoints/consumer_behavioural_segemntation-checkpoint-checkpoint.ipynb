# Consumer Behavior & Segmentation (Behavior-Based)
## Business Question

## Which consumer segments drive growth and profitability?

# Import Libraries

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Load Data

```python
# load data
df = pd.read_parquet("fmcg_cleaned.parquet")
```

# SKU-Level Behavioural Metrics

```python
sku_behavior = (
    df.groupby('sku')
      .agg(
          avg_price=('price_unit','mean'),
          promo_ratio=('promotion_flag','mean'),
          total_volume=('units_sold','sum'),
          total_revenue=('revenue','sum')
      )
      .reset_index()
)

# Segemnt logic
def assign_segment(row):
    if row['promo_ratio'] > 0.6:
        return 'Promo Driven'
    elif row['avg_price'] > sku_behavior['avg_price'].median():
        return 'Premium'
    elif row['total_volume'] > sku_behavior['total_volume'].median():
        return 'Core Mass'
    else:
        return 'Price Sensitive'

sku_behavior['consumer_segment'] = sku_behavior.apply(assign_segment, axis=1)

# segment Contribution 
segment_perf = (
    sku_behavior.groupby('consumer_segment')['total_revenue']
                .sum()
                .sort_values(ascending=False)
)

segment_perf.plot(kind='bar', title='Revenue by Consumer Segment')
plt.show()
```
