# Price, Promotion & Supply Constraints

## Business Question

## How do pricing and promotions influence demand and performance?

## Import Required Libraries

```python
# import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyarrow
```

## Load Data

```python
# load data
df = pd.read_parquet("fmcg_cleaned.parquet")
```

## Promotion Lift

```python
# Promotion lift
promo_lift = (
    df.groupby('promotion_flag')['units_sold']
      .mean()
)

promo_lift.plot(kind='bar', title='Promo vs Non-Promo Average Units Sold')
plt.show()
```

## Delivery and Stock Impact

```python
# Delivery and stock impact
stock_impact = (
    df.groupby('out_of_stock_flag')['units_sold']
      .mean()
)

stock_impact.plot(kind='bar', title='Impact of Stock Availability on Sales')
plt.show()


```

## Delivery Speeds vs Sales

```python
delivery_sales = (
    df.groupby('delivery_days')['units_sold']
      .mean()
)

delivery_sales.plot(title='Average Units Sold vs Delivery Time')
plt.show()

```

## Insights
### Promotions fail when stock or fulfillment constraints are ignored.
